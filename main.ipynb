```python
from textblob import TextBlob
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate sentiment polarity
def get_sentiment(text):
    if pd.isna(text):
        return 0  # Treat missing text as neutral
    return TextBlob(text).sentiment.polarity

# Add a new column to the dataset for sentiment polarity
data['SentimentPolarity'] = data['post_body_text'].apply(get_sentiment)

# Group the data by sentiment categories (positive, neutral, negative) based on polarity
data['SentimentCategory'] = pd.cut(data['SentimentPolarity'], bins=[-1, -0.1, 0.1, 1], labels=['Negative', 'Neutral', 'Positive'])

# Calculate average engagement metrics for each sentiment category
sentiment_engagement = data.groupby('SentimentCategory')[['LikesCount', 'SharesCount', 'CommentsCount']].mean().reset_index()

# Plotting the sentiment analysis results
plt.figure(figsize=(10, 6))
bar_width = 0.25
x = np.arange(len(sentiment_engagement['SentimentCategory']))

plt.bar(x, sentiment_engagement['LikesCount'], width=bar_width, label='Likes', color='green')
plt.bar(x + bar_width, sentiment_engagement['SharesCount'], width=bar_width, label='Shares', color='orange')
plt.bar(x + 2 * bar_width, sentiment_engagement['CommentsCount'], width=bar_width, label='Comments', color='blue')

# Adding labels and title
plt.xlabel('Sentiment Category')
plt.ylabel('Average Count')
plt.title('Average Engagement by Sentiment of Posts')
plt.xticks(x + bar_width, sentiment_engagement['SentimentCategory'])
plt.legend(title='Engagement Type')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
# Split the data into features and target labels

X_train = manual_shan['post_body_text']  # Replace with your feature extraction method
y_train = manual_shan['encoded_labels']

from sklearn.feature_extraction.text import TfidfVectorizer
# Create a TF-IDF matrix
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=218)
X_tfidf_trained = tfidf_vectorizer.fit_transform(X_train)

# send the output of the vectorizer into a dataframe
tfidf_df = pd.DataFrame(X_tfidf_trained.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df

# Train Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(X_tfidf_trained, y_train)
```

```python
# Example: Assuming `unlabeled_data` is a DataFrame of new texts
X_unlabeled=df['post_body_text']
X_tfidf_unlabeled = tfidf_vectorizer.fit_transform(X_unlabeled)
# Predict labels for the unlabeled data
predicted_labels = lr_model.predict(X_tfidf_unlabeled)
X_unlabeled = df['post_body_text']

# If you want probabilities instead of hard labels, use predict_proba
predicted_probabilities = lr_model.predict_proba(X_tfidf_unlabeled)

# Add the predictions back to the DataFrame
df['predicted_label'] = predicted_labels
df
```
